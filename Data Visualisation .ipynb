```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
y5 = np.arcsin(x) # sin-1
y6 = np.arccos(x) # cos-1


#(rows,column)

fig,axs = plt.subplots(3,2, figsize = (10,10))

axs[0,0].plot(x,y1,color = 'red', label = "Sin(X)")
axs[0,0].set_title("Sin Graph")
#axs[0,0].legend()
axs[0,0].set_xlabel("X Axis")
axs[0,0].set_ylabel("Y Axis")
axs[0,0].grid()



axs[0,1].plot(x,y2, color = "blue", label = "Cos(X)")
axs[0,1].set_title("Cos Graph")
axs[0,1].legend()
axs[0,1].set_xlabel("X Axis")
axs[0,1].set_ylabel("Y Axis")





axs[1,0].plot(x,y3, color = "green", label = "Tan(X)")
axs[1,0].set_title("Tan Graph")
axs[1,0].legend()
axs[1,0].set_xlabel("X Axis")
axs[1,0].set_ylabel("Y Axis")





axs[1,1].plot(x,y4,color ="purple", label = "Exp(X)")
axs[1,1].legend()

axs[2,0].plot(x,y5,color ="red", label = "ArcSin(X)")
axs[2,0].legend()


axs[2,1].plot(x,y6,color ="skyblue", label = "ArcCos(X)")
axs[2,1].legend()





plt.tight_layout()
plt.title("Mathematical Graph")
plt.legend()
plt.show()
```
